Introduction
The goal of this task is to explore and visualize the Iris flower dataset, a classic dataset in machine learning. It contains 150 records of iris flowers, categorized into three species: Setosa, Versicolor, and Virginica. Each record includes four features: sepal length, sepal width, petal length, and petal width.

In this task, we will:

Load and inspect the Iris dataset

Summarize the structure and key statistics

Create visualizations like scatter plots, histograms, and box plots

Understand relationships and distributions in the data

```python
# Task 1: Exploring and Visualizing the Iris Dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nDataset info:")
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
df['petal_length'].hist(bins=20, color='Red', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

Conclusion
In this task, we successfully explored the Iris dataset and gained meaningful insights through visualization and analysis. Key observations include:

Setosa flowers are visually distinct and easily separable based on petal size.

The petal length and width are highly effective in distinguishing between species.

No missing data was found, and the dataset is clean and well-structured.

Box plots revealed that Versicolor and Virginica may have overlapping features, making them harder to separate than Setosa.
